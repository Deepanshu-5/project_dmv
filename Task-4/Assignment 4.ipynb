## Object-Oriented Programming (OOP)

Object-Oriented Programming is a programming paradigm that organizes software around objects rather than functions alone.
    The primary goal of OOP is to improve:

1) Modularity 
2) Reusability
3) Maintainability
4) Scalability

Instead of managing data and functions separately, OOP encapsulates them into logical units that represent real-world or system entities.

## Theory notes

### Class
A class is a blueprint or template used to create objects.

A class defines:
* Attributes (data)
* Methods (behavior)
  
A class itself does not occupy meaningful business state.
It only describes what an object should look like.


```python
class Student:
    pass
```

### Object
An object is an instance of a class.

When a class is instantiated, memory is allocated and an object is created.

Each object maintains its own state while sharing behavior defined by the class.

```python
class user:
    pass

u1 = user()
print(type(u1))
```

### Constructor (init)

A constructor is a special method automatically executed when an object is created.

Purpose:

* Initialize object state
* Ensure required data exists
* Establish object consistency

```python
class user:
    def __init__(self, name):
        self.name = name

u1 = user("Deepanshu")
print(u1.name)
```

### Instance Variables and Methods
Instance variables belong to an object and store its data. Instance methods operate on the object's data and use the self parameter to access instance variables

```python
class User:

    def __init__(self, name):
        self.name = name

    def display(self):
        print(self.name)

u1 = User("Deepanshu")
u1.display()
```

### Class Variables

Class variables belong to the class itself.
They are shared among all objects.

```python
class Student:

    school = "ABC School"

print(Student.school)
```

### Encapsulation

Encapsulation is the practice of restricting direct access to internal object data and controlling modifications through methods.
Goals:

* Data protection
* Validation
* Controlled access

### Inheritance

Inheritance allows one class to acquire properties and behaviors from another class.

Purpose:

* Code reuse
* Hierarchical relationships

 python uses MRO(method resolution order )to find method

```python
class Parent:

    def show(self):
        print("Parent")

class Child(Parent):
    pass

c = Child()
c.show()
```

### Polymorphism

Polymorphism allows different objects to respond differently to the same method call.

```python
class Car:
    def move(self):
        return "Drive!"

class Boat:
    def move(self):
        return "Sail!"


def start_journey(vehicle):
    print(vehicle.move())

start_journey(Car())   
start_journey(Boat())
```

### Abstraction

Abstraction hides implementation details while exposing only necessary functionality.

Users interact with what a system does rather than how it works internally.abstract class cannot be instantiated b/c they contain incomplete behaviour and serve only as contracts for subclasses

```python
from abc import ABC, abstractmethod

class Shape(ABC):

    @abstractmethod
    def area(self):
        pass
```

### Method Overriding

Method overriding occurs when a child class provides its own implementation of a method already defined in the parent class.

Purpose:

* Specialize behavior
* Preserve common interface

```python
class Parent:

    def show(self):
        print("Parent Method")

class Child(Parent):

    def show(self):
        print("Child Method")

Child().show()
```

### super() Function

super() allows a child class to access methods and constructors from its parent class.

Purpose:

* Reuse parent functionality
* Avoid code duplication
* Maintain inheritance chains

```python
class Parent:

    def __init__(self):
        print("Parent Constructor")

class Child(Parent):

    def __init__(self):
        super().__init__()
        print("Child Constructor")

Child()
```

### Access Modifiers
Python supports Public, Protected, and Private members. Public members can be accessed anywhere. Protected members use a single underscore. Private members use double underscores and are intended to be hidden.

```python
class Demo:

    public = 10
    _protected = 20
    __private = 30

print(Demo.public)
```

## 2. Bank Account System

```python
class Bankaccount:

    def __init__(self, balance):

        if not isinstance(balance, (int, float)):
            raise TypeError("Balance must be numeric")

        if balance < 0:
            raise ValueError("Initial balance cannot be negative")

        self.__balance = balance

    def deposit(self, amount):

        if not isinstance(amount, (int, float)):
            raise TypeError("Deposit amount must be numeric")

        if amount <= 0:
            raise ValueError("Deposit amount must be greater than zero")

        self.__balance += amount

        print(f"Deposited ₹{amount}")

    def withdraw(self, amount):

        if not isinstance(amount, (int, float)):
            raise TypeError("Withdrawal amount must be numeric")

        if amount <= 0:
            raise ValueError("Withdrawal amount must be greater than zero")

        if amount > self.__balance:
            raise ValueError("Insufficient balance")

        self.__balance -= amount

        print(f"Withdrawn ₹{amount}")

    def get_balance(self):
        return self.__balance

account = Bankaccount(10000)
print(f"the current balance is  : ₹{account.get_balance()}")
account.deposit(3000)
account.withdraw(5000)
print(f"the current balance is  : ₹{account.get_balance()}")
```

## 5.  Shape Area Calculator (Abstraction)

```python
from abc import ABC, abstractmethod
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass
        
class Rectangle(Shape):
    def __init__(self, length, breadth):
        self.length = length
        self.breadth = breadth

    def area(self):
        return self.length * self.breadth

class Circle(Shape):
    def __init__(self,radius):
        self.radius=radius
    def area(self):
        return 3.14*self.radius*self.radius

rect=Rectangle(10,5)
print(f"the area is : {rect.area()}")
c=Circle(7)
print(f"the area is : {c.area()}")
```

## 6. Library Management System (Mini Project)

```python
class Book:
    
    def __init__(self,book_id,title,author):
        self.book_id=book_id
        self.title=title
        self.author=author
        self.borrowed_by=None
        
    def marked_borrowed(self,member_id):
        if not isinstance(member_id, int):
            raise TypeError("invalid member_id")
        if self.borrowed_by is not None:
            raise ValueError("sorry book is borrowed already")
        self.borrowed_by=member_id

    def marked_returned(self):
        if self.borrowed_by is None:
            raise ValueError("book is already returned ")
        self.borrowed_by=None
        
    def is_available(self):
      return self.borrowed_by is None

class Member:
   
    def __init__(self,name,member_id):
        self.name=name
        self.member_id=member_id

class Library:
    def __init__(self):
   
        self.books = {}
        self.members = {}

    def add_book(self, book):
      if book.book_id in self.books:
         raise ValueError("Book ID already exists")
      self.books[book.book_id] = book
        
    def add_member(self, member):
      if member.member_id in self.members:
         raise ValueError("Member ID already exists")
      self.members[member.member_id] = member

    def issue_book(self, book_id, member_id):
        if book_id not in self.books:
          raise ValueError("Book not found")
        if member_id not in self.members:
          raise ValueError("Member not found")
        book = self.books[book_id]
        book.marked_borrowed(member_id)
    
    def return_book(self, book_id):
        if book_id not in self.books:
          raise ValueError("Book not found")
        book=self.books[book_id]
        book.marked_returned()
  
    def display_available_books(self):
        print("\nAvailable Books:\n")
        for book in self.books.values():
         if book.is_available():
            print(
                f"ID: {book.book_id} | "
                f"Title: {book.title} | "
                f"Author: {book.author}"
            )
book1 = Book(1, "Python", "James")
book2 = Book(2, "AI", "Harish")

member1 = Member("Sakshi", 101)
member2 = Member("Lokesh", 102)

lib = Library()
lib.add_book(book1)
lib.add_book(book2)
lib.add_member(member1)
lib.add_member(member2)

lib.display_available_books()
lib.issue_book(1, 101)
lib.display_available_books()
lib.return_book(1)
lib.display_available_books()
```
