# Functions & Code Reusability Assignment

## Assignment 1 — BMI Calculator

### Theory Notes

### Functions

Functions are reusable blocks of code that perform specific tasks.

Advantages:
- Reusability
- Better readability
- Easier debugging
- Modular programming

### Parameters and Arguments

- Parameters are variables in function definition.
- Arguments are actual values passed during function call.

### Return Statement

The `return` keyword sends output back to the caller.

```python
print("BMI CALCULATOR")
def calculate_bmi(weight, height):
    bmi = weight / (height ** 2)
    return bmi

w = float(input("enter your weight in kg"))
h = float(input("enter your height in meter"))
```

```python
def get_bmi_category(bmi):
    if bmi < 18.5:
        return "Underweight"
    elif bmi < 25:
        return "Normal Weight"
    elif bmi < 30:
        return "Overweight"
    return "Obese"
    
if h>0 and w>0:
  bmi = calculate_bmi(w,h)
  print("BMI:", round(bmi, 2))
  category = get_bmi_category(bmi)
  print("Category:", category)
else:
    print("invalid input")
```

### Explanation

- `calculate_bmi()` calculates Body Mass Index.
- `get_bmi_category()` classifies BMI into categories.
- Functions improve modularity and code reuse.
- `return` sends data back from the function.

## Assignment 2 — EMI Calculator

### Theory Notes

### EMI

EMI stands for Equated Monthly Installment.

It is the fixed monthly payment made by a borrower to repay a loan.

### Functions

Functions improve:
- modularity
- reusability
- readability
- maintainability

### Return Statement

The `return` keyword sends computed data back to the caller.

### EMI Formula

$$
EMI = \frac{P \times R \times (1+R)^N}{(1+R)^N - 1}
$$

Where:

- P = Principal Loan Amount
- R = Monthly Interest Rate
- N = Number of Monthly Installments

```python
def calculate_emi(principal, annual_rate, years):
    monthly_rate = annual_rate / (12 * 100)
    months = years * 12
    numerator = principal * monthly_rate * ((1 + monthly_rate) ** months)
    denominator = ((1 + monthly_rate) ** months) - 1
    emi = numerator / denominator
    return emi

P=float(input("enter principal amount in rupees"))
R=float(input("enter your annual rate :"))
T=float(input("enter your time period of emi in years :"))
if P>0 and T>0 and R>0:
    emi = calculate_emi(P, R,T)
    print("Monthly EMI: Rupees", round(emi, 2),"per month")
```

### Explanation

- `calculate_emi()` calculates the monthly loan installment.
- The function uses the EMI mathematical formula.
- Parameters make the function reusable.
- `return` sends EMI value back to the caller.

## Assignment 3 — Recursive Factorial Program

### Theory Notes

#### Recursion

Recursion is a process where a function calls itself.

A recursive function must contain:
- Base condition
- Recursive call

#### Factorial

Factorial of a number is the product of all positive integers less than or equal to the number.

Example:

5! = 5 × 4 × 3 × 2 × 1 = 120

```python
def factorial(n):
    if n == 1 or n==0:
        return 1
    return n * factorial(n - 1)

num=int(input("enter a number please : "))
if num>=0:
   result = factorial(num)
   print("Factorial:", result)
else:
  print("inavlid input ")
```

### Explanation

- `factorial()` is a recursive function.
- The function calls itself repeatedly.
- The base condition stops infinite recursion.
- Recursion improves problem decomposition.

### Complexity Analysis

#### Time Complexity

O(n)

#### Space Complexity

O(n)

Reason:
- Each recursive call uses stack memory.

## Assignment 4 — Fibonacci Series Generator

### Theory Notes

#### Fibonacci Series

The Fibonacci sequence is a series where each number is the sum of the previous two numbers.

Example:

0, 1, 1, 2, 3, 5, 8, 13 ...

#### Functions

Functions improve:
- modularity
- code reuse
- readability

#### Recursion

Recursion occurs when a function calls itself.

```python
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)

def generate_fibonacci_series(limit):
    for i in range(limit):
        print(fibonacci(i), end=" ")

num=int(input("enter a number for fibonacci series"))
if num>0:
  generate_fibonacci_series(num)
else:
    print("please enter number greater than 0")
```




The recursive Fibonacci algorithm generates numbers where each number is the sum of the previous two numbers.
Formula:

$$
F(n) = F(n-1) + F(n-2)
$$

Base Conditions:

$$
F(0) = 0
$$

$$
F(1) = 1
$$

#### Recursion

Recursion occurs when a function calls itself repeatedly until a base condition is met.

#### Advantages

- Cleaner mathematical representation
- Easier to understand recursion concepts
- Demonstrates divide-and-conquer computation

#### Disadvantages

- Repeated calculations
- Higher memory usage
- Slower execution for large inputs

### Complexity Analysis

#### Time Complexity

$$
O(2^n)
$$

Reason:
- Each function call creates two additional recursive calls.
- Many Fibonacci values are recalculated multiple times.

#### Space Complexity

$$
O(n)
$$

Reason:
- Recursive calls use stack memory.
- Maximum recursion depth depends on `n`.

#### Optimization Insight

Recursive Fibonacci is inefficient for large inputs.

Professional systems optimize this using:
- Memoization
- Dynamic Programming
- Iterative Computation

## Assignment 5 — Prime Number Checker

### Theory Notes

#### Prime Number

A prime number is a number greater than 1 that has only two factors:
- 1
- itself

Examples:
- 2
- 3
- 5
- 7

#### Functions

Functions improve:
- modularity
- reusability
- maintainability

#### Optimization

Efficient algorithms reduce unnecessary computations.

```python
def is_prime(number):

    if number <= 1:
        return False

    for i in range(2, int(number ** 0.5) + 1):

        if number % i == 0:
            return False

    return True


number=int(input("Enter a number for checking "))
if is_prime(number):
    print(number, "is a Prime Number")
else:
    print(number, "is not a Prime Number")
```

### Explanation

- `is_prime()` checks whether a number is prime.
- The function returns `True` or `False`.
- Optimization is performed using square root logic.
- Functions improve code reuse and readability.

### Complexity Analysis

#### Time Complexity

$$
O(\sqrt{n})
$$

Reason:
- The loop runs only till the square root of the number.

#### Space Complexity

$$
O(1)
$$

Reason:
- No extra memory is used except variables.

#### Optimization Insight

Checking divisibility till `n` is inefficient.

Using square root optimization significantly reduces computations.

## Assignment 6 — List Operations using map(), filter(), and reduce()

### Theory Notes

#### map()

`map()` applies a function to every element in an iterable.

#### filter()

`filter()` selects elements based on a condition.

#### reduce()

`reduce()` repeatedly applies a function to reduce iterable values into a single result.

#### Functional Programming

Functional programming focuses on:
- reusable functions
- transformation pipelines
- declarative computation

```python
from functools import reduce
```

### map() Example

```python
numbers = [1, 2, 3, 4, 5]

squared_numbers = list(map(lambda x: x ** 2, numbers))

print("Squared Numbers:", squared_numbers)
```

### filter() Example

```python
numbers = [1, 2, 3, 4, 5, 6, 7, 8]

even_numbers = list(filter(lambda x: x % 2 == 0, numbers))

print("Even Numbers:", even_numbers)
```

### reduce() Example

```python
numbers = [1, 2, 3, 4, 5]

total = reduce(lambda x, y: x + y, numbers)

print("Sum:", total)
```

### Explanation

- `map()` transforms data.
- `filter()` selects data based on conditions.
- `reduce()` aggregates iterable values.
- Lambda functions provide compact reusable logic.
- Functional programming improves readability and modularity.
